Task 1: Creating Tensors

```python
import torch

list_tensor = torch.tensor([1, 2, 3, 4, 5])

zeros_tensor = torch.zeros((3, 3))

ones_tensor = torch.ones((2, 4))

rand_tensor = torch.rand((3, 3))

identity_tensor = torch.eye(3)

float_tensor = torch.tensor([1.5, 2.5], dtype=torch.float32)
int_tensor = torch.tensor([1, 2], dtype=torch.int64)

print("List tensor:", list_tensor)
print("Zeros:\n", zeros_tensor)
print("Ones:\n", ones_tensor)
print("Random:\n", rand_tensor)
print("Identity:\n", identity_tensor)
print("Float tensor dtype:", float_tensor.dtype)
```

Task 2: Tensor Information

```python
t = torch.rand((3, 4, 2))

print("Shape (Shakli):", t.shape)
print("Data type (Ma'lumot turi):", t.dtype)
print("Device (Qurilma - CPU/GPU):", t.device)
print("Number of dimensions (O'lchamlar soni):", t.ndim)
```

Task 3: Tensor Operations

```python
a = torch.tensor([[1.0, 2.0], [3.0, 4.0]])
b = torch.tensor([[5.0, 6.0], [7.0, 8.0]])

add_res = a + b

mul_res = a * b

matmul_res = torch.matmul(a, b) # yoki a @ b

reshaped = a.reshape(1, 4)

flattened = a.flatten()

concat_res = torch.cat((a, b), dim=0)

print("Addition:\n", add_res)
print("Multiplication:\n", mul_res)
print("Matrix Multiplication:\n", matmul_res)
print("Reshaped:\n", reshaped)
print("Flattened:\n", flattened)
print("Concatenated:\n", concat_res)
```

Task 4: Indexing and Slicing

```python
grid = torch.tensor([[10, 20, 30],
                     [40, 50, 60],
                     [70, 80, 90]])

single_elem = grid[1, 1] # 50

first_row = grid[0, :] # Birinchi satr

second_col = grid[:, 1] # Ikkinchi ustun

sub_grid = grid[0:2, 1:3]

grid[0, 0] = 999

print("Single Element:", single_elem.item())
print("First Row:", first_row)
print("Second Column:", second_col)
print("Sliced Grid:\n", sub_grid)
print("Modified Grid:\n", grid)
```

Task 5: NumPy and Tensors

```python
import numpy as np

np_array = np.array([1, 2, 3, 4, 5])
tensor_from_np = torch.from_numpy(np_array)

tensor_data = torch.ones(3, 3)
np_from_tensor = tensor_data.numpy()

print("Tensor from NumPy:", tensor_from_np)
print("NumPy from Tensor:\n", np_from_tensor)
```

Task 6: Mini Challenge

```python
challenge_tensor = torch.rand(4, 5)

tensor_mean = challenge_tensor.mean()
tensor_max = challenge_tensor.max()
tensor_min = challenge_tensor.min()
tensor_sum = challenge_tensor.sum()

print("Random 4x5 Tensor:\n", challenge_tensor)
print(f"Mean: {tensor_mean.item():.4f}")
print(f"Max: {tensor_max.item():.4f}")
print(f"Min: {tensor_min.item():.4f}")
print(f"Sum: {tensor_sum.item():.4f}")
```

What is a tensor?

Tenzor — chuqur o'rganishda ishlatiladigan ko'p o mezonli sonlar matritsasi bo'lib, u skalyar ($0$D), vektor ($1$D) va matritsa ($2$D) kabi tushunchalarni yanada yuqori o'lchamlarga umumiylashtiruvchi ma'lumot tuzilmasidir.

Why are tensors important in deep learning?

Tenzorlar barcha turdagi ma'lumotlarni (rasmlar, matnlar, audiolar) neyron tarmoqlari tushunadigan shaklda saqlash va ular ustida GPU yordamida tezkor parallel matematik amallar bajarish imkonini bergani uchun muhim.
